```python
# Search Trends
from pytrends.request import TrendReq

pytrends = TrendReq(hl="en-US", tz=360)

kw_list = ["artificial intelligence", "ChatGPT", "Claude", "generative AI", "machine learning"]
pytrends.build_payload(
    kw_list,
    timeframe="2022-01-01 2025-01-01",
    geo="US"
)

df = pytrends.interest_over_time()

print(df.tail(20))
```

```python
from bs4 import BeautifulSoup
import re

with open(r"E:\Om\signalscope\data\sec\sec-edgar-filings\NVDA\10-K\0001045810-22-000036\full-submission.txt",
          "r",
          encoding="utf-8",
          errors="ignore") as f:
    raw = f.read()

# Extract only the 10-K document text
text_match = re.search(r"<TEXT>(.*?)</TEXT>", raw, re.DOTALL | re.IGNORECASE)
html = text_match.group(1) if text_match else raw

# Remove HTML/XBRL tags
soup = BeautifulSoup(html, "html.parser")
clean_text = soup.get_text(" ")

# Normalize spaces
clean_text = re.sub(r"\s+", " ", clean_text).lower()

keywords = [
    "artificial intelligence",
    "generative ai",
    "machine learning",
    "ai",
    "data center",
    "accelerated computing"
]

for kw in keywords:
    print(kw, clean_text.count(kw))
```
